```python
import pandas as pd
from typing import Dict, List
import matplotlib.pyplot as plt
```

```python
# Create the DataFrame

# parameters for this cell
filepath: str = './../wifi_data_until_20190204.csv'
time_col_name = 'time'
time_col_index: int = 0
occupancy_type = 'int32'
occupancy_na = -1

data: pd.DataFrame = pd.read_csv(
    filepath_or_buffer=filepath,
    # time column
    index_col=time_col_index,
    parse_dates=[time_col_index],
    infer_datetime_format=True,
    # debugging
    nrows=50
).fillna(occupancy_na)
    
# All columns of the DataFrame should be of an unsigned integer type
for name, series in data.iteritems():
    # Occupancy count can only be nonnegative integers.
    # Occupancy in a building will never exceed 2^32.
    if name != time_col_name:
        data[name] = series.astype(dtype=occupancy_type, copy=False)
```

```python
print(data.dtypes)
print(data.columns)
```

```python
# Draw Line Graph Matrix

# Parameters
chart_rows = 3
chart_cols = 3
plt.style.use('seaborn-darkgrid')
palette = plt.get_cmap('Set1')

num = 0
for i in range(1, min(chart_rows*chart_cols, len(data.columns))):
    
    num += 1

    col_name = data.columns[i]
 
    plt.subplot(chart_rows, chart_cols, num)

    # Plot the lineplot
    plt.plot(
        data[time_col_index],
        data[i],
        marker='',
        color=palette(num),
        linewidth=1.9,
        alpha=0.9,
        label=col_name
    )

    
    plt.xlim(data.index.min, data.index.max)
    plt.ylim(min(data[i]), max(data[i]))
 
    # Not ticks everywhere
    if num in range(7) :
        plt.tick_params(labelbottom='off')
    if num not in [1,4,7] :
        plt.tick_params(labelleft='off')
```
